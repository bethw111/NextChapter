```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("middlelight/goodreadsbookswithgenres")

print("Path to dataset files:", path)
```

```python
import os
files = os.listdir(path)
print(files)
```

```python
import pandas as pd

df = pd.read_csv(os.path.join(path, "Goodreads_books_with_genres.csv"))

df.head()
```

```python
df.columns
```

```python
df = df.dropna(subset=["Title", "Author", "num_pages", "genres"])
```

```python
def clean_genre(g):
    return g.split(",")[0].strip().lower()

df["genre_clean"] = df["genres"].apply(clean_genre)
```

```python
def map_length(pages):
    if pages < 150:
        return "Short"
    elif pages < 300:
        return "Medium"
    else:
        return "Long"
    
df["length"] = df["num_pages"].apply(map_length)
```

```python
df = df.drop_duplicates(subset=["isbn"], keep='first')
```

```python
df.describe()
```

```python
df.to_csv("cleaned_books.csv", index=False)
```

```python
df["text"] = (df["Title"] + " " + df["Author"] + " " + df["genres"])
```
